```python
race_ids = races[races.year==2023].raceId.tolist()

race_ids = [race_ids[-1]]

r23 = results[results.raceId.isin(race_ids)]
print(race_ids)

r23.head()

driver_list = r23.driverId.unique().tolist()

final_standings = []

for did in driver_list:
    final_standings.append(r23[r23.driverId==did])

final_standings = pd.concat(final_standings)

final_standings = final_standings.sort_values(by='points', ascending=False)
# final_standings = final_standings.iloc[::-1]

cols = ['position', 'points', 'raceId', 'wins', 'number', 'driverId', 'driverRef', 'code', 'forename', 'surname', 'url']

res = final_standings.loc[:,cols]
```

```python
import logging
import math
import pandas as pd
from pathlib import Path
import dash
from dash import html, Input, Output, callback, State, dcc, dash_table, clientside_callback, ctx
from dash.exceptions import PreventUpdate

col_titles = html.Thead(
    html.Tr([
        html.Th('Points', scope='col'),
        html.Th('First', scope='col'),
        html.Th('Last', scope='col'),
        html.Th('Code', scope='col'),
        html.Th('Wins', scope='col'),
    ]),
    className='div-conta
)

col_body = []

for row in res.itertuples():
    position = row[1]
    points = row[2]
    raceId = row[3]
    wins = row[4]
    number = row[5]
    driverId = row[6]
    driverRef = row[7]
    code = row[8]
    forename = row[9]
    surname = row[10]
    url = row[11]

    col_body.append(
        html.Tr([
            html.Th(points, scope="row"),
            html.Td(f"{forename} {surname}"),
            html.Td(f"{code}"),
            html.Td(f"{wins}"),
        ],
            className="table-active"
        )
    )



return html.Table([col_titles, html.Tbody(col_body)], className="table table-hover")
    
```
